```python
### PARAMETERS

### MultiIndex level slice constant:
All = slice(None)
### NA for MS Excel files:
list_na_excel_values = ['', '#N/A', '#N/A N/A', '#NA', '-1.#IND', '-1.#QNAN', '-NaN', '-nan', '1.#IND', '1.#QNAN', 'N/A', 'NULL', 'NaN', 'n/a', 'nan', 'null',
                        '#N/A Requesting Data...', '#N/A Invalid Security', '#N/A Field Not Applicable']
### Industry mapping path:
str_path_industry_map = 'Data_Files/Source_Files/WIOT_to_GICS_mapping.xlsx'
str_sheet_mapping = 'WIOT Mapping'
str_sheet_full_gics = 'GICS List Old'
### Localization path:
str_path_localization = 'Data_Files/Source_Files/Localization.xlsx'
### Path to original WIOT Tables:
str_path_wiot_source = 'Data_Files/Source_Files/WIOT'
### Downloaded values:
str_path_gics_hdf = 'Data_Files/Result_Files/old_way_flows_value.h5'
### Aggregated shares:
str_path_region_distrib_hdf = 'Data_Files/Result_Files/old_way_regional_to_regional_share.h5'
### Aggregated shares:
str_path_region_weight_hdf = 'Data_Files/Result_Files/old_way_regional_to_regional_weight.h5'
### Universal key:
str_gics_key = 'gics_io'
```

```python
### WIOD TO GICS INDUSTRY MAPPING

### SIC to GICS:
dict_industry_mapper = pd.read_excel(io = str_path_industry_map, sheet_name = str_sheet_mapping, header = [0], index_col = 0)\
                         .squeeze().astype(str).to_dict()
### GISC Group substitutions (to filling the gaps):
dict_gics_sub = pd.read_excel(io = str_path_industry_map, sheet_name = str_sheet_full_gics, header = [0], index_col = None, dtype = str)\
                  .replace({np.NaN: None}).set_index('Industry Group').squeeze().to_dict()
```

```python
### DEFINING COUNTRY CODES EXTRACTOR

def get_country_codes(use_local_copy = False):  
    ### In case if URL is unavailable:
    if (use_local_copy):
        url_country_code = 'Data_Files/Source_Files/countrycode.html'
    ### Online extraction:
    else:
        url_country_code = 'https://countrycode.org/'
    df_full_codes = pd.read_html(url_country_code, index_col = 'COUNTRY')[0]
    df_full_codes[['ISO SHORT', 'ISO LONG']] = df_full_codes['ISO CODES'].str.split(' / ', expand = True)
    df_result = df_full_codes[['ISO SHORT', 'ISO LONG']].sort_index()    
    df_result.index = df_result.index.str.upper()
    ### Results output:
    return df_result

### World Country Codes:
df_country_codes = get_country_codes()
dict_ison_mapper = df_country_codes.set_index('ISO LONG').squeeze().to_dict()
dict_ison_mapper['ROW'] = 'YY'
```

```python
### TO REPLACE WITH SQL REQUEST: EXTRACTION AF ACTUAL ISON MEMBERS LIST

def ison_membership_converting(str_path_universe, date_end, bool_daily = False, int_backfill_months = 0):
    ### Defining business-month-end reindexation on country level:
    def country_modify(ser_raw_country, date_end):
        ser_res_country = ser_raw_country.droplevel(0).resample('MS').last().resample('BM').last()
        range_country = pd.date_range(ser_res_country.index[0], date_end, freq = 'BM')
        return ser_res_country.reindex(range_country).ffill()
    ### Markets encoding table:
    dict_markets = {50 : 'DM', 57 : 'EM', 504 : 'FM', 0: np.NaN}     
    ### Loading source file:
    df_raw_universe = pd.read_excel(io = str_path_universe, sheet_name = 'Switchers', header = 0, parse_dates = True, date_format='%m/%d/%Y', index_col = [0, 1],
                                 na_values = list_na_excel_values, keep_default_na = False)
    ### Converting source file:
    df_raw_universe.index.names = ['Country', 'Date']
    ser_raw_universe = df_raw_universe['Region']
    ser_raw_universe.fillna(0, inplace = True)
    ser_raw_universe.name = 'Market'
    ### By country reindexation and translation:
    ser_res_universe = ser_raw_universe.groupby('Country').apply(country_modify, date_end)
    ser_res_universe.index.names = ['Country', 'Date']
    ser_res_universe = ser_res_universe.replace(dict_markets).reorder_levels([1, 0]).sort_index() 
    ### Expanding membership for primary regions members by backfilling:
    if int_backfill_months:
        ### List of regions:
        list_region = list(ser_res_universe.dropna().unique())
        ### Initialising of collection of series with backfilled data for each region:
        list_ison_backfill = []
        ### Regions looping:
        for iter_region in list_region:
            ### Defining start of region date:
            date_first_valid = ser_res_universe.loc[ser_res_universe == iter_region].first_valid_index()[0]
            ### Creating dates index to backfilling:
            idx_date_backfill = pd.date_range(end = date_first_valid, periods = int_backfill_months + 1, freq = 'BM')[: -1]
            ### Creating primary countries index to backfilling:            
            idx_region_backfill = ser_res_universe.loc[ser_res_universe == iter_region].loc[date_first_valid, All].index.get_level_values('Country')
            ### Creating full index:
            idx_ison_backfill = pd.MultiIndex.from_product([idx_date_backfill, idx_region_backfill])
            ### Series with backfilled data:
            list_ison_backfill.append(pd.Series(iter_region, index = idx_ison_backfill))
        ### Combination of backfilled series and original ISON data:    
        ser_res_universe = ser_res_universe.combine_first(pd.concat(list_ison_backfill, axis = 0)).sort_index()  
        ser_res_universe.index.names = ['Date', 'Country']
    ### Converting to daily frequency:
    if bool_daily:
        ser_res_universe = ser_res_universe.reset_index('Country').groupby('Country').resample('B').ffill()['Market'].swaplevel().sort_index()    
    ### Results output:
    ser_res_universe.name = 'Market'
    return ser_res_universe

### NA for MS Excel files:
list_na_excel_values = ['', '#N/A', '#N/A N/A', '#NA', '-1.#IND', '-1.#QNAN', '-NaN', '-nan', '1.#IND', '1.#QNAN', 'N/A', 'NULL', 'NaN', 'n/a', 'nan', 'null',
                        '#N/A Requesting Data...', '#N/A Invalid Security', '#N/A Field Not Applicable', '---']
### Universe path:
str_path_universe = 'Data_Files/Source_Files/acadian_universe.xlsx'
### ISON membership history:
ser_ison_membership = ison_membership_converting(str_path_universe, pd.to_datetime('2023-12-29'))
### ISON membership status:
ser_ison_status = ser_ison_membership[ser_ison_membership.index.levels[0][-1]]
```

```python
### DATA TABLES LOADING & CONVERTATION

dict_flows = {}
#for year_matrix_csv in os.listdir(str_path_wiot_source):
for year_matrix_csv in os.listdir(str_path_wiot_source)[::-1]:    
#for year_matrix_csv in [os.listdir(str_path_wiot_source)[0]] + [os.listdir(str_path_wiot_source)[-1]]:
    gc.collect()
#    dt_year_end = (pd.to_datetime(year_matrix_csv[4 : 8]) + pd.offsets.BYearEnd()).date()
    str_year_num = year_matrix_csv[4 : 8]
    print(str_year_num)
    df_year_raw = pd.read_csv(str_path_wiot_source + '/' + year_matrix_csv, sep = ';', index_col = [0, 1, 2, 3], skiprows = [0, 1, 2, 3], header = [0, 1], 
                                        na_values = list_na_excel_values, keep_default_na = False)
    ser_year_raw = df_year_raw.droplevel([0, 1]).stack([0, 1]).astype('int32')
    ser_year_raw.name = 'Value'
    ser_year_raw.index.names = ['Supplier_Country_Long', 'Supplier_Industry_r_named', 'User_Country_Long', 'User_Industry_c_named']
    df_year_typed = ser_year_raw.reset_index()
    del df_year_raw
    del ser_year_raw
    gc.collect()
    df_year_typed = df_year_typed[df_year_typed['Supplier_Country_Long'].isin(df_country_codes['ISO LONG'].to_list() + ['ROW']) & 
                                  df_year_typed['User_Country_Long'].isin(df_country_codes['ISO LONG'].to_list() + ['ROW'])]
    df_year_typed['Supplier_Country'] = df_year_typed['Supplier_Country_Long'].replace(dict_ison_mapper)
    df_year_typed['User_Country'] = df_year_typed['User_Country_Long'].replace(dict_ison_mapper)    
    df_year_typed['User_Industry_r_named'] = df_year_typed['User_Industry_c_named'].str.replace('c', 'r')
    df_year_typed = df_year_typed[df_year_typed['Supplier_Industry_r_named'].isin(dict_industry_mapper.keys()) & 
                                  df_year_typed['User_Industry_r_named'].isin(dict_industry_mapper.keys())]    
    df_year_typed['User_Industry'] = df_year_typed['User_Industry_r_named'].replace(dict_industry_mapper)
    df_year_typed['Supplier_Industry'] = df_year_typed['Supplier_Industry_r_named'].replace(dict_industry_mapper)
    df_year_typed['Supplier_Country'] = df_year_typed['Supplier_Country'].astype('category')
    df_year_typed['User_Country'] = df_year_typed['User_Country'].astype('category')
    df_year_typed['User_Industry'] = df_year_typed['User_Industry'].astype('category')
    df_year_typed['Supplier_Industry'] = df_year_typed['Supplier_Industry'].astype('category')
    df_year_typed = df_year_typed[['Supplier_Country', 'User_Country', 'Supplier_Industry', 'User_Industry', 'Value']]
    ser_year_gics = df_year_typed.groupby(['Supplier_Country', 'User_Country', 'Supplier_Industry', 'User_Industry']).sum().squeeze().sort_index()
#    ser_flow_share = ser_year_gics.groupby(['Supplier_Country', 'User_Country', 'Supplier_Industry'])\
#                                  .apply(lambda ser_group: ser_group / ser_group.sum()).sort_index()
#    ser_flow_share  = ser_flow_share.astype('float32')
    dict_flows[str_year_num] = ser_year_gics    

#    break
```

```python
### CONVERTED DATA AGGREGATION & SAVING

gc.collect()

ser_flows = pd.concat(dict_flows, names = ['Date'])
ser_flows.to_hdf(str_path_gics_hdf, key = str_gics_key, mode = 'w', format = 'table', complevel = 9)
```

```python
### REGIONAL (AS SUPPLIER) & REGIONAL (AS USER) VALUES CONSTRUCTION

dict_region_share = {}
dict_region_weight = {}

for iter_year in range(2000, 2015)[::-1]:
    gc.collect()
    print(iter_year)    
    str_iter_year = str(iter_year)
    ### Flows extraction:
    df_iter_flows = pd.read_hdf(str_path_gics_hdf, key = str_gics_key, where = "Date = str_iter_year").reset_index().drop('Date', axis = 1)
    df_iter_flows = df_iter_flows[(df_iter_flows['Supplier_Industry'] != '9999') & (df_iter_flows['User_Industry'] != '9999')] 
    ### Information adding:
    ser_ison_status.loc['YY'] = 'FM'
    ser_ison_status.name = 'Supplier_Region'    
    df_iter_flows = df_iter_flows.merge(ser_ison_status, left_on = 'Supplier_Country', right_index = True, how = 'inner')
    ser_ison_status.name = 'User_Region'    
    df_iter_flows = df_iter_flows.merge(ser_ison_status, left_on = 'User_Country', right_index = True, how = 'inner')
#    ### Adding CN as a separate User Region:
#    df_iter_flows_CN = pd.DataFrame(df_iter_flows.loc[df_iter_flows['User_Country'].isin(['CN'])].values, columns = df_iter_flows.columns)
#    df_iter_flows_CN['User_Region'] = 'CN'
#    df_iter_flows = pd.concat([df_iter_flows, df_iter_flows_CN], axis = 0, ignore_index = True)
    ### Supply regions aggregation:
    dict_iter_regions_agg = {}
    dict_iter_regions_agg['DM'] = df_iter_flows[df_iter_flows['Supplier_Region'] == 'DM']\
                                                .groupby(['User_Region', 'Supplier_Industry', 'User_Industry'], observed = True)['Value'].sum()
    dict_iter_regions_agg['EM'] = df_iter_flows[df_iter_flows['Supplier_Region'] == 'EM']\
                                                .groupby(['User_Region', 'Supplier_Industry', 'User_Industry'], observed = True)['Value'].sum()
    dict_iter_regions_agg['FM'] = df_iter_flows[df_iter_flows['Supplier_Region'] == 'FM']\
                                                .groupby(['User_Region', 'Supplier_Industry', 'User_Industry'], observed = True)['Value'].sum()
    dict_iter_regions_agg['CN'] = df_iter_flows[df_iter_flows['Supplier_Country'] == 'CN']\
                                                .groupby(['User_Region', 'Supplier_Industry', 'User_Industry'], observed = True)['Value'].sum()
#    dict_iter_regions_agg['EM_ex_CN'] = df_iter_flows[(df_iter_flows['Supplier_Region'] == 'EM') & (df_iter_flows['Supplier_Country'] != 'CN')]\
#                                                .groupby(['User_Localization', 'Supplier_Industry', 'User_Industry'], observed = True)['Value'].sum() 
    ser_iter_regions_agg = pd.concat(dict_iter_regions_agg, names = ['Supplier_Region'])
    ### Adding regions to world data:
#    ser_iter_regions_to_world = pd.concat([ser_iter_regions_agg.loc[:, ['DM', 'EM', 'FM']].groupby(['Supplier_Region', 'Supplier_Industry', 'User_Industry'], 
#                                           observed = True).sum()], keys = ['World'], names = ['User_Region'])
    ser_iter_regions_to_world = pd.concat([ser_iter_regions_agg.groupby(['Supplier_Region', 'Supplier_Industry', 'User_Industry'], observed = True).sum()],\
                                          keys = ['World'], names = ['User_Region'])    
    ser_iter_regions_to_world = ser_iter_regions_to_world.swaplevel(0, 1).sort_index()    
    ser_iter_regions_agg = pd.concat([ser_iter_regions_agg, ser_iter_regions_to_world], axis = 0).sort_index()
    ### Dropping Industry internal trade inside the Region:
    df_iter_regions_agg = ser_iter_regions_agg.reset_index()
    df_iter_regions_agg.loc[(df_iter_regions_agg['Supplier_Region'] == df_iter_regions_agg['User_Region']) & 
                            (df_iter_regions_agg['Supplier_Industry'] == df_iter_regions_agg['User_Industry']), 'Value'] = 0
    ser_iter_regions_agg = df_iter_regions_agg.set_index(['Supplier_Region', 'User_Region', 'Supplier_Industry', 'User_Industry']).squeeze().sort_index()    
    ### Distribution calculation:
    ser_iter_regions_shares = ser_iter_regions_agg.groupby(['Supplier_Region', 'User_Region', 'Supplier_Industry'], observed = True)\
                                                  .transform(lambda ser_group: ser_group if (ser_group.sum() == 0) else ser_group / ser_group.sum())\
                                                  .astype('float32').sort_index()  
    dict_region_share[str_iter_year] = ser_iter_regions_shares 
    ### Weights calculation:
    ser_iter_regions_weights = ser_iter_regions_agg.drop('World', axis = 0, level = 'User_Region')\
                                                   .groupby(['Supplier_Region', 'User_Region', 'Supplier_Industry'], observed = True).sum()\
                                                   .groupby(['Supplier_Region', 'Supplier_Industry'], observed = True)\
                                                   .transform(lambda ser_group: ser_group if (ser_group.sum() == 0) else ser_group / ser_group.sum())\
                                                   .astype('float32').sort_index()
    dict_region_weight[str_iter_year] = ser_iter_regions_weights
#    break
```

```python
### REGIONAL WEIGHTS VISUALIZATION

def draw_barplot(ser_group):
    ser_group.name = 'Weight'
    str_region = ser_group.index[0][0]
    df_group = ser_group.droplevel('Supplier_Region').reset_index()
    plt.figure(figsize = (12, 4))
    ax_histplot = sns.histplot(data = df_group, x = 'Supplier_Industry', hue = 'User_Region', multiple = 'fill', weights = 'Weight', discrete = True, shrink = 0.8)

    ax_histplot.set(xlabel = 'Industry Group of Supplier')
    ax_histplot.set(ylabel = 'Weight of Localization in Industry Supply')
    ax_histplot.set_title('Regional Distribution : ' + str_region + ' by User Localization', pad = 10)
    plt.xticks(rotation = 45)
    
    plt.show()

ser_region_weight = pd.read_hdf(str_path_region_weight_hdf)
ser_region_last = ser_region_weight['2014']
ser_region_last = ser_region_last.unstack(['Supplier_Region', 'User_Region']).reindex(dict_gics_sub.keys())\
                                 .stack(['Supplier_Region', 'User_Region'], dropna = False).squeeze().sort_index()
df_region_last = ser_region_last.unstack(['Supplier_Region', 'User_Region'])
for iter_industry in dict_gics_sub:
    if (dict_gics_sub[iter_industry] is not None):
        str_iter_sub = dict_gics_sub[iter_industry]
#        print(iter_industry, '/', str_iter_sub)
        df_region_last.loc[iter_industry] = df_region_last.loc[str_iter_sub].values
ser_region_last = df_region_last.stack(['Supplier_Region', 'User_Region'], dropna = False).reorder_levels([1, 2, 0]).sort_index()
ser_region_last.name = 'Weight'

ser_region_last.groupby('Supplier_Region', group_keys = True, sort = True).apply(draw_barplot)
```

```python
### LOADING REGION TO WORLD SHARES 2014 & SAVING DATA TO CSV

ser_region_share = pd.read_hdf(str_path_region_distrib_hdf)
ser_region_last = ser_region_share['2014']

def region_filling(ser_iter_region):  
    ser_iter_region = ser_iter_region.droplevel(['Supplier_Region', 'User_Region']).unstack('Supplier_Industry').reindex(dict_gics_sub.keys()).stack(dropna = False)\
                                     .unstack('User_Industry').reindex(dict_gics_sub.keys()).stack(dropna = False).sort_index()
    for iter_industry in dict_gics_sub:
        if (dict_gics_sub[iter_industry] is not None):
            str_iter_sub = dict_gics_sub[iter_industry]
            ser_iter_region.loc[iter_industry] = ser_iter_region[str_iter_sub].values
    ser_iter_region.fillna(0.0, inplace = True)
    return ser_iter_region

ser_region_last = ser_region_last.groupby(['Supplier_Region', 'User_Region'], group_keys = True, sort = True).apply(region_filling)
ser_region_last.unstack('User_Industry').to_csv('Data_Files/Result_Files/old_way_regional_to_regional_share.csv', sep = ',')
```
